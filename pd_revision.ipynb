```python
# Data transformation

# removing duplicate

#  "del df['column']" command for removing column and "drop (row)" command for row

df = pd.DataFrame({'color':['white', 'white','red', 'yellow', 'white' ],
                    'value':[1,2,3,4,2]})

df.duplicated()

# Extract duplicats

df[df.duplicated()]

# to removing duplicates

df.drop_duplicates()

```

```python
# Replacing values via mapping

# replace(), map(), rename()

df1 = pd.DataFrame({
    'scrip':['sbi','infy', 'tcs','coalindia'],
    'sector':['baks', 'it', 'it company', 'energy']
})

corrected_sector = {
    "baks":"banking",
    "it company":"it"
}

df1.replace(corrected_sector)

```

```python
# adding values via mapping

df2 = pd.DataFrame({
    'scrip':['infy','sbi','tcs'],
    'sector':['it','banking', 'it']
})

close_price = {
    'infy':1600,
    'tcs':3300,
    'sbi':450
}

df2['close price']=df2['scrip'].map(close_price)

df2
```

```python
reindex={
    0:'first',
    1:"second",
    2:"third"
}

# renaming index

df3 = df2.rename(reindex)

df3
```

```python
#  for column renaming using column argument

recolumn = {
    'scrip':"ticker"
}

df5 = df3.rename(index=reindex, columns=recolumn)

df5

# to retain changes within the dataframe set inplace option to true

df3.rename(index=reindex, columns=recolumn, inplace=True)

df3
```

```python
df = pd.DataFrame({
    'color':['red','green','blue', 'red','green','blue'],
    'item':['ball','pen','ball', 'pen', 'pencil','ball'],
    'price1':np.random.random(6),
    'price2':np.random.random(6),
})

# Data Aggregation

grp = df['price1'].groupby(df['color'])

grp.mean()

grp.sum()
```

```python
# Hierachial grouping

grp=df['price1'].groupby([df['color'],df['item']])

grp.groups

df1=grp.mean()

# df1.unstack()

```

```python
df[['price1','price2']].groupby(df['color']).mean()
```

```python
# Group Iteration

for name, group in df.groupby('color'):

    print(name)
    print(group)
```

```python
# Chain transformation
r1 = df['price1'].groupby(df['color']).mean()
print(r1)
type(r1)
```

```python
r2 = df.groupby(df['color']).mean() # entire dataframe is grouped here 

print(type(r2))
r2
```

```python
means = df.groupby(df['color']).mean().add_prefix('mean_')

means
```

```python
# Functions on groups

g = df.groupby('color')
g['price1'].quantile(0.5)
```
