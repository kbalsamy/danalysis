Data preparation - Assembling -> {merging, concatenating, combining}

```python
# merging

frame1 = pd.DataFrame( {'id':['ball','pencil','pen','mug','ashtray'], 'price': [12.33,11.44,33.21,13.23,33.62]})
frame1
```

```python
frame2 = pd.DataFrame( {'id':['pencil','pencil','ball','pen'], 'color': ['white','red','red','black']})
frame1, frame2
```

```python
pd.merge(frame1, frame2)
```

```python
frame1 = pd.DataFrame( {'id':['ball','pencil','pen','mug','ashtray'],
'color': ['white','red','red','black','green'],
'brand': ['OMG','ABC','ABC','POD','POD']})

frame2 = pd.DataFrame( {'id':['pencil','pencil','ball','pen'],
'brand': ['OMG','POD','ABC','POD']})

frame1, frame2 
```

```python
pd.merge(frame1, frame2, on='id')
```

```python
pd.merge(frame1, frame2, on='brand')
```

```python
frame2.columns=['sid','brand']
frame2.columns
frame2
```

```python
pd.merge(frame1, frame2, left_on='id', right_on='sid')
```

```python
display(frame1)
display(frame2)

frame2.columns=['id', 'brand']
pd.merge(frame1, frame2, on="id", how='outer')
```

```python
pd.merge(frame1,frame2, left_index=True, right_index=True)
```

```python
results = [12,34,67,55,28,90,99,12,3,56,74,44,87,23,49,89,87]
print(len(results))
bins = [0,25,50,75,100]

results, bins

bin = pd.cut(results, bins)

bin,bin.categories, bin.codes

# print(pd.value_counts(bin))

bin_names = ['unlikely', 'less likely','likely', 'highly likely' ]

named_bin=pd.cut(results, bins, labels=bin_names)

named_bin
```
