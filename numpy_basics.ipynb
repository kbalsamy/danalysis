we can reshape the array using np.reshape(array to be reshaped, newshape=(), order="C/F/A")

```python
array_to_reshape = np.arange(10)
array_to_reshape.reshape(2,5)
```

Alternate way 

```python
np.reshape(array_to_reshape,newshape=(2,5), order="C")
```

Adding new axis to an array / convert 1D array to 2D array
np.newaxis() and np.expand_dims() methods handle conversion

np.expand_dims(), using this method, new dimentions can be inserted at given index i.e. row or column

```python
test_array = np.arange(10)

convert_2d_array = test_array[np.newaxis, :]

add_dim_to_row=convert_2d_array.shape

print(add_dim_to_row)

add_dim_to_col= test_array[:,np.newaxis]

print(add_dim_to_col.shape)
```

creating one dimensional array then converting it to a 2 dimensional array then slice and select

```python
test= np.linspace(0,20, num=12) 
test1 = np.reshape(test,newshape=(3,4))
print(test1)

test2 = test1[(test1>10)]

test3 = test1[(test1==14) | (test1>=14)]

test3

test4 = (test1==14) | (test1>=14)

test4
```
